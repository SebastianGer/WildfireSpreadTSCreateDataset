```python
import numpy as np
from Preprocessing.PreprocessingService import PreprocessingService
preprocessing = PreprocessingService()
```

```python
import datetime
data_dir = '/Users/zhaoyu/PycharmProjects/CalFireMonitoring/data/SwedishFire'
start_date = datetime.date(2018,7,17)
delta = 15
```

```python
from pathlib import Path
from glob import glob
import matplotlib.pyplot as plt
import cv2
from imageio import imsave
for i in range(delta):
    data_dir_with_date = Path(data_dir + '/' + str(start_date + datetime.timedelta(days=i)))
    file_list = glob(str(data_dir_with_date / "*.tiff"))
    file_list.sort()
    f1, _ = preprocessing.read_tiff(file_list[0])
    s2, _ = preprocessing.read_tiff(file_list[1])
    s3, _ = preprocessing.read_tiff(file_list[2])
    s6, _ = preprocessing.read_tiff(file_list[3])
    f1 = f1.transpose((1,2,0))
    s2 = s2.transpose((1,2,0))
    s3 = s3.transpose((1,2,0))
    s6 = s6.transpose((1,2,0))
                      
    nbr = (s6-s3)/(s6+s3)
    print('F1 image '+str(start_date + datetime.timedelta(days=i)))
    plt.imshow(f1, cmap='gray')
    plt.show()
    ret, f1_th = cv2.threshold(f1, 320,10000, cv2.THRESH_TOZERO)
    print('F1 thresholded image '+str(start_date + datetime.timedelta(days=i)))
    plt.imshow(f1_th, cmap='gray')
    plt.show()
    imsave(data_dir+'/f1_thresholded'+str(start_date + datetime.timedelta(days=i))+'.png', f1_th)
    
    print('NBR image '+str(start_date + datetime.timedelta(days=i)))
    plt.imshow(nbr, cmap='gray')
    plt.show()
    ret, nbr_th = cv2.threshold(nbr, -0, 1, cv2.THRESH_TOZERO)
    print('NBR thresholded image '+str(start_date + datetime.timedelta(days=i)))
    plt.imshow(nbr_th, cmap='gray')
    plt.show()
    imsave(data_dir+'/nbr_thresholded' + str(start_date + datetime.timedelta(days=i)) + '.png', nbr_th)
```
